# Data Cleaning Products Dataset

##1. Create a Copy of the Dataset

```python
products_df = products.copy()
```

##2. Remove Duplicate Records

```python
products_df = products_df.drop_duplicates()
```

```python
products_df.duplicated().sum()
```

All duplicate rows were successfully removed from the dataset

##3. Review Dataset Information

```python
products_df.info()
```

##4. Handle Missing Values

```python
(products.isna().mean() * 100).round(2)
```

```python
products_df.isna().sum()
```

All rows containing missing values were successfully removed

##5. Correct Data Types

```python
products_df['price'].str.count(r"\.").value_counts()
```

```python
# @title Default title text
mult_decimal_rows = (products_df["price"].str.count(r"\.") > 1).sum()

percent_corrupted = (100 * mult_decimal_rows / products_df.shape[0])

print(f"{percent_corrupted:.2f}% of the rows in the products dataset have multiple decimal points in the price column.")
```

The price column contains 373 values with multiple decimal points, representing 3.56% of the dataset. Since these values cannot be correctly converted into a numeric format and represent only a small proportion of the data, these rows will be removed before converting the column to a numeric data type.

```python
products_df = products_df[products_df["price"].str.count(r"\.") <= 1]
```

```python
products_df.shape
```

```python
products_df["price"]
```

```python
products_df["price"] = pd.to_numeric(products_df["price"])
```

```python
products_df.dtypes
```

After removing the invalid values, the price column was successfully converted from object to float64

```python
products_df['promo_price'].str.count(r"\.").value_counts()
```

```python
products_df = products_df.drop(columns=["promo_price"])
```

The products dataset was successfully cleaned.

```python
products_df.to_csv(PATH + "products_clean.csv", index=False)
```
